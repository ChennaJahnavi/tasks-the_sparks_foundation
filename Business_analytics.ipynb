```
df.drop(['Country'],axis=1,inplace=True)
```

```
df.columns
```

```
numerical_features=[]
for i in df.columns:
  if df[i].dtype!='O':
    numerical_features.append(i)
```

```
numerical_features
```

```
categorical_features=[]
for i in df.columns:
  if df[i].dtype=='O':
    categorical_features.append(i)
```

```
categorical_features
```

```
for feature in numerical_features:
  data=df.copy()
  data.groupby(feature)['Profit'].mean().plot.bar(figsize=(12,7))
  plt.xlabel(feature,fontsize=2)
  plt.ylabel('Profit')
  plt.title(feature)
  plt.show()
```

here we could see that postal code doesn't effect profit much

```
df.columns
```

```
for feature in categorical_features:
  data=df.copy()
  data.groupby(feature)['Profit'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Profit')
  plt.title(feature)
  plt.show()
```

```
for feature in numerical_features:
  data=df.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    plt.xlabel(feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

```
corr = df.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
